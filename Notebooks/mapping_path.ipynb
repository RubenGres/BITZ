```python
import json
import folium

with open("history_example.json", "r") as file:
    history = json.load(file)

messages = history["history"]

gps_coordinates = [message["image_location"] for message in messages if "image_location" in message]
```

```python
valid_coords = [coord for coord in gps_coordinates if coord]
valid_coords[:3]
```

```python
center_lat, center_lon = valid_coords[0]["latitude"], valid_coords[0]["longitude"]
m = folium.Map(location=[center_lat, center_lon], zoom_start=15)

for coord in valid_coords:
    folium.Marker([coord["latitude"], coord["longitude"]]).add_to(m)

m
```
